## Area Pots (Series/Dataframe) <a id="12"></a>

**What is a line plot and why use it?**

An Area chart or area plot is a type of plot which displays information as a series of data points called 'markers' connected by straight line segments. It is a basic type of chart common in many fields. Use line plot when you have a continuous data set. These are best suited for trend-based visualizations of data over a period of time.

**Questions:**

Buatlah Area plot untuk menunjukkan tren perubahan dalam jumlah kasus kejahatan dari waktu ke waktu. Area plot dapat membantu Anda untuk mengidentifikasi tren musiman atau tren jangka panjang dalam kejahatan.

```python
# Group by total crimes berdasarkan borough dan year lalu reset index
df_borough = df.groupby(["borough","year"])["value"].sum().reset_index(name="total_crime")
# Set borough as index and year component as a column using pivot
df_borough = df_borough.pivot(index='borough', columns='year', values='total_crime').reset_index()
# Save to excel
df_borough.to_excel('london_crime_by_lsoa.xlsx', index=False)
```

```python
from pandas import ExcelWriter

#Baca data dari excel
df_new = pd.read_excel('london_crime_by_lsoa.xlsx')
# Set borough sebagai index
df_new = df_new.set_index("borough")
# Sum total_crime dari tahun 2008 sampai 2016
df_new['total_crime'] = df_new.loc[:, 2008:2016].sum(axis=1)
# Sort data berdasarkan total_crime descending
df_new.sort_values(['total_crime'], ascending=False, axis=0, inplace=True)
# Kenakan transpose
df_new.transpose()
# Drop total_cime
df_new2 = df_new.drop("total_crime", axis=1)
# Ambil top 5
df_new2_top5 = df_new2.head(5)
# Transpose kembali
df_new2_top5 = df_new2_top5.transpose()
```

```python
# Buat area plot
#Plot dalam area plot
df_new2_top5.plot(kind='area', stacked=False, alpha=0.55, figsize=(20, 10), cmap='inferno')

# Set nilai minimum sumbu Y
plt.ylim(25000, 60000)

# Add title and label
plt.title("Jumlah Kejahatan di London Periode 2008 - 2016")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Kejahatan")

# Tampilkan plot
plt.show()
```

**Insight:**

Borough Westminster ialah borough dengan jumlah kejatan yang paling banyak dari periode 2008 hingga 2016. Dapat dilihat pada borough Westminster dan borough lainnya bahwa pada periode 2012 terjadi lonjakan kejahatan dibanding tahun sebelum dan sesudahnya

## Histogram

A histogram is a way of representing the frequency distribution of numeric dataset. The way it works is it partitions the x-axis into bins, assigns each data point in our dataset to a bin, and then counts the number of data points that have been assigned to each bin. So the y-axis is the frequency or the number of data points in each bin. Note that we can change the bin size and usually one needs to tweak it so that the distribution is displayed nicely.

**Question:**

Tunjukan distribusi frekuensi jumlah kejahatan dari seluruh borough di London pada periode 2008 - 2016.

```python
borough_unique = df["borough"].unique()
jumlah_borough = len(borough_unique)

print(f"Jumlah borough yang berbeda: {jumlah_borough}")
```

```python
# Write your function below
# bin intervals
count, bin_edges = np.histogram(df_new['total_crime'])
# Buat histogram untuk total tindakan kriminal diseluruh borough di London pada 2008 - 2016
# Gunakan warna yang kontras
plt.hist(df_new['total_crime'], color='#29ABCA')

# Tambahkan judul dan label
plt.title('Histogram dari Jumlah Total Tindak Kriminal di 33 Borough Pada Periode 2008 - 2016')
plt.ylabel('Jumlah Borough')
plt.xlabel('Jumlah Tindak Kriminal')

# Tambahkan garis pembatas bin
plt.grid(which='major', color='lightgray', linestyle='--', linewidth=0.5)

# Tambahkan garis rata-rata
plt.axvline(df_new['total_crime'].mean(), color='red', linestyle='--')

plt.show()
```

**Insight:**

Berdasarkan histogram di atas. Terdapat kurang lebih 12 borough yang memiliki 20000 tindak kriminal selama periode 2008 - 2016. Persebaran tindak kriminal merata antara borough lainnya.

## Bar Charts (Dataframe) <a id="10"></a>

A bar plot is a way of representing data where the *length* of the bars represents the magnitude/size of the feature/variable. Bar graphs usually represent numerical and categorical variables grouped in intervals. 

To create a bar plot, we can pass one of two arguments via `kind` parameter in `plot()`:

* `kind=bar` creates a *vertical* bar plot
* `kind=barh` creates a *horizontal* bar plot

**Question:**

Westminster merupakan daerah dengan tindakan kriminal paling banyak di Kota London dari periode 2008 - 2016. Dengan periode terbanyak terjadinya tindak kriminal ialah pada tahun 2011. Analisalah persebaran dan jumlah tindak kriminal yang paling sering terjadi di Westminster Borough pada tahun 2011.

```python
# Write your function below
# Filter data untuk borough Westminster dan tahun 2011
df_filter = df.loc[(df["borough"] == "Westminster") & (df["year"] == 2011)]
# Kelompokan persebaran crime dan jumlahnya di borough Westminster pada tahun 2011
df_crime = df_filter.groupby('major_category', sort=False)['value'].sum()

# Buat plot bar
plt.figure(figsize=(15, 7))
plt.bar(df_crime.index, df_crime.values)

# Tambahkan judul dan label sumbu
plt.title('Berbagai Macam Tindak Kriminal dan Jumlahnya di Westminster Borough Pada Tahun 2011', fontsize=16, weight='bold')
plt.xlabel('Tindak Kriminal', fontsize=14, weight='bold')
plt.ylabel('Jumlah', fontsize=14, weight='bold')

# Miringkan label di sumbu-x
plt.xticks(rotation=45)

# Tampilkan plot
plt.show()
```

**Insight:**

Dapat kita lihat bahwa tindak kriminal paling banyak di Westminster pada 2011 ialah Theft and Handling berjumlah 30000++ kasus. Sedangkan, tidak kriminal lainnya ialah bejumlah 1000 - 8000 kasus.

## Pie Charts <a id="6"></a>

A `pie chart` is a circualr graphic that displays numeric proportions by dividing a circle (or pie) into proportional slices. You are most likely already familiar with pie charts as it is widely used in business and media. We can create pie charts in Matplotlib by passing in the `kind=pie` keyword.

**Question:**

Buatlah analisa mengenai presentasi tindak kriminal paling banyak di Kota London dari periode tahun 2008 hingga 2016

```python
# Write your function below
# Menghitung persebaran tindakan kriminal di kota London
df_crm_ldn = df.groupby('major_category', sort=False)['value'].sum().reset_index()

# ratio for each continent with which to offset each wedge.
explode_list = [0.05, 0.05, 0.05, 0.025, 0.025, 0.05, 0.025, 0.025, 0.1]

# Membuat pie chart
df_crm_ldn['value'].plot(kind='pie',
                figsize=(12, 6),
                autopct='%1.1f%%',  # menampilkan persentase dengan 1 digit desimal
                startangle=90,
                shadow=True,
                labels=None,  # menonaktifkan label pada pie chart
                pctdistance=1.12,  # jarak teks persentase dari pie chart
                explode=explode_list,  # meledakkan irisan tertentu
                colors=['lightcoral', 'skyblue', 'lightgreen', 'lightyellow', 'magenta', 'orange', 'mediumpurple', 'gold', 'teal'],  # custom warna irisan
                )

# scale the title up by 12% to match pctdistance
plt.title('Persentase Persebaran Tindak Kriminal di London', y=1.12)  # sesuaikan judul sesuai kebutuhan

plt.axis('equal')  # membuat pie chart berbentuk lingkaran sempurna

# tambahkan legend
plt.legend(labels=df_crm_ldn['major_category'], loc='upper left')  # gunakan label kategori kriminal dari DataFrame

plt.show()
```

**Insight:**

Presentasi tindak kriminal di kota London pada periode 2008 hingga 2016 ialah theft and handling sebanyak 41.3% diikuti oleh violence against person 24.2%

## Box Plots <a id="8"></a>

A `box plot` is a way of statistically representing the *distribution* of the data through five main dimensions: 

- **Minimun:** Smallest number in the dataset.
- **First quartile:** Middle number between the `minimum` and the `median`.
- **Second quartile (Median):** Middle number of the (sorted) dataset.
- **Third quartile:** Middle number between `median` and `maximum`.
- **Maximum:** Highest number in the dataset.

**Question:**

Identifikasi distribusi nilai pada data tindak kriminal Theft and Handling di Westminster pada tahun 2011

```python
# Write your function below
# Filter data untuk kategori Theft and Handling di borough Westminster pada tahun 2011
df_filter2 = df.loc[(df["borough"] == "Westminster") & (df["year"] == 2011) & (df["major_category"] == "Theft and Handling")]
# Kelompokan persebaran crime dan jumlahnya di borough Westminster pada tahun 2011
df_filter2 = df_filter2.groupby("month")["value"].sum().reset_index()
# Atur month menjadi index
df_filter2.set_index("month", inplace = True)

# Plot Boxplot
df_filter2.plot(kind='box', figsize=(7, 6))

plt.title('Boxplot Tindak Kriminal "Theft and Handling" di Westminster Setiap Bulannya Pada Tahun 2011')
plt.ylabel('Jumlah tindak kriminal')
plt.xlabel('Theft and Handling')

plt.show()
```

```python
df_filter2.describe()
```

**Insight:**

Distribusi nilai pada kategori tindak kriminal Theft and Handling pada tahun 2011 memiliki nilai min: 2300, Q1: 2500, Median: 2700, Q3: 3000, dan max: 3400. Terdapat data yang diduga sebagai outlier bernilai 3600 diluar batas atas dari boxplot

## Scatter Plots <a id="10"></a>

A `scatter plot` (2D) is a useful method of comparing variables against each other. `Scatter` plots look similar to `line plots` in that they both map independent and dependent variables on a 2D graph. While the datapoints are connected together by a line in a line plot, they are not connected in a scatter plot. The data in a scatter plot is considered to express a trend. With further analysis using tools like regression, we can mathematically calculate this relationship and use it to predict trends outside the dataset.

**Question:**

Carilah hubungan antara jumlah tindakan kriminal dengan nilai gdp dari setiap borough di London pada periode 2008 - 2016

```python
#Tambahkan kolom baru 'gdp' di df_new yang berisikan random integer
df_new['gdp'] = np.random.randint(1000, 10000, size=len(df_new))

# Plot scatter plot
plt.scatter(df_new['total_crime'], df_new['gdp'])

# Tambahkan garis tren
plt.plot([df_new['total_crime'].min(), df_new['total_crime'].max()],
     [np.poly1d(np.polyfit(df_new['total_crime'], df_new['gdp'], 1))(df_new['total_crime'].min()),
     np.poly1d(np.polyfit(df_new['total_crime'], df_new['gdp'], 1))(df_new['total_crime'].max())],
     color='red')

# Tambahkan garis kisi
plt.grid(which='major', color='lightgray', linestyle='--', linewidth=0.5)

# Tambahkan label dan judul
plt.xlabel('Jumlah Tindak Kriminal')
plt.ylabel('GDP')
plt.title('Grafik Scatter Plot Jumlah Tindak Kriminal vs GDP dari setiap Borough di London')

plt.show()
```

**Insight:**

Garis tren pada grafik scatter plot tidak menunjukkan adanya hubungan antara jumlah tindak kriminal dengan gdp dari setiap borough di London

## Word Clouds <a id="8"></a>


`Word` clouds (also known as text clouds or tag clouds) work in a simple way: the more a specific word appears in a source of textual data (such as a speech, blog post, or database), the bigger and bolder it appears in the word cloud.

```python
# install wordcloud
# !conda install -c conda-forge wordcloud --yes

!pip install wordcloud

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
stopwords = set(STOPWORDS)
```

```python
# instantiate a word cloud object
df_wc = df.groupby('minor_category')['value'].sum().sort_values(ascending=False).reset_index(name='total_criminal')

# instantiate a word cloud object
sum_wc = dict(zip(df_wc['minor_category'], df_wc['total_criminal']))

your_wordcloud = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)
# generate the word cloud
your_wordcloud.generate_from_frequencies(sum_wc)
```

```python
# Plot wordcloud
fig = plt.figure()
fig.set_figwidth(12) # set width
fig.set_figheight(14) # set height

# display the cloud
plt.imshow(your_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## Folium

Folium is a powerful Python library that helps you create several types of Leaflet maps. The fact that the Folium results are interactive makes this library very useful for dashboard building.

From the official Folium documentation page:

> Folium builds on the data wrangling strengths of the Python ecosystem and the mapping strengths of the Leaflet.js library. Manipulate your data in Python, then visualize it in on a Leaflet map via Folium.

> Folium makes it easy to visualize data that's been manipulated in Python on an interactive Leaflet map. It enables both the binding of data to a map for choropleth visualizations as well as passing Vincent/Vega visualizations as markers on the map.

> The library has a number of built-in tilesets from OpenStreetMap, Mapbox, and Stamen, and supports custom tilesets with Mapbox or Cloudmade API keys. Folium supports both GeoJSON and TopoJSON overlays, as well as the binding of data to those overlays to create choropleth maps with color-brewer color schemes.

```python
#!conda install -c conda-forge folium=0.5.0 --yes
!pip install folium
import folium

print('Folium installed and imported!')
```

```python
# define the world map centered around Semarang with a low zoom level
london_map = folium.Map(location=[51.509865, -0.118092], zoom_start=12)
incidents = folium.map.FeatureGroup()
incidents.add_child(
        folium.CircleMarker(
            [51.509865, -0.118092],
            radius=5,
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add pop-up text to each marker on the map
folium.Marker([51.509865, -0.118092], popup='WOW').add_to(london_map)

# add incidents to map
london_map.add_child(incidents)

# display world map
london_map
```
