Now we can analyze Confidence Intervals below.

Function below are graded function. (1 Points)

```python
def calculate_confidence_interval(newdf, confidence=0.95):
    """
    Calculate the confidence interval for the mean number of residential units.
    
    Parameters:
    data (Series or array-like): Input data containing the number of residential units.
    confidence (float): Confidence level for the interval (default is 0.95).
    
    Returns:
    tuple: Lower and upper bounds of the confidence interval.
    """
    confidence_interval = stats.norm.interval(confidence, loc=newdf.mean(), scale=newdf.std())
    return confidence_interval

# Calculate confidence interval for the mean number of residential units
confidence_interval = calculate_confidence_interval(newdf['RESIDENTIAL UNITS'])

# Print the confidence interval
print("Confidence Interval for the Mean Number of Residential Units:", confidence_interval)
```

Make your Hypothesis Testing below

Function below are graded function. (1 Points)

We want to test whether the mean sale price of properties in Brooklyn (Borough 3) is significantly different from the mean sale price of properties in Manhattan (Borough 1). By comparing the mean sale prices, we can assess whether there are substantial differences in real estate prices between Brooklyn and Manhattan. This information is valuable for various purpose.

Hypotheses:

Null Hypothesis (H0): The mean sale price of properties in Brooklyn is equal to the mean sale price of properties in Manhattan.

Alternative Hypothesis (H1): The mean sale price of properties in Brooklyn is not equal to the mean sale price of properties in Manhattan.

```python
# Graded-Function Begin
def hypothesis_testing_sale_price(newdf, borough1, borough2):
    """
    Perform hypothesis testing to compare the mean sale prices between two boroughs.
    
    Parameters:
    data (DataFrame): Input DataFrame containing the property sales data.
    borough1 (int): Borough code for the first borough.
    borough2 (int): Borough code for the second borough.
    
    Returns:
    tuple: t-statistic and p-value from the hypothesis test.
    """
    # Filter data for the two boroughs
    borough1_data = newdf[newdf['BOROUGH'] == borough1]['SALE PRICE']
    borough2_data = newdf[newdf['BOROUGH'] == borough2]['SALE PRICE']
    
    # Remove missing or invalid values
    borough1_data = borough1_data.dropna()
    borough2_data = borough2_data.dropna()
    
    # Perform two-sample t-test if both boroughs have valid data
    if len(borough1_data) > 0 and len(borough2_data) > 0:
        t_statistic, p_value = stats.ttest_ind(borough1_data, borough2_data, equal_var=False)
        return t_statistic, p_value
    else:
        return None, None

# Define the borough codes for Brooklyn and Manhattan
borough1 = 3  # Brooklyn
borough2 = 1  # Manhattan

# Perform hypothesis testing
t_statistic, p_value = hypothesis_testing_sale_price(newdf, borough1, borough2)

# Print the results
print("Results of Hypothesis Testing:")
print("Borough 1 (Brooklyn) Code:", borough1)
print("Borough 2 (Manhattan) Code:", borough2)
if t_statistic is not None and p_value is not None:
    print("t-statistic:", t_statistic)
    print("p-value:", p_value)
else:
    print("Unable to perform hypothesis testing due to missing or invalid data.")
# Graded-Function End
```

Write your final conclusion below.

Your conclusion below are graded. (1 Points)

The p-value, which is practically zero, indicates an extremely strong statistical significance. In simpler terms, this means that the likelihood of observing such a large difference in mean sale prices between Brooklyn and Manhattan by pure chance is nearly impossible.

Therefore, we confidently reject the idea that the mean sale prices between Brooklyn and Manhattan are the same. In other words, there's a real, tangible difference in property values between the two boroughs.

Moreover, the negative t-statistic indicates that, on average, properties in Brooklyn tend to sell for significantly less than properties in Manhattan. This difference is not just a minor fluctuation but a substantial gap in pricing between the two areas
